```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data_df["SN"].unique())
total_player_df = pd.DataFrame({"Total Players": [player_count]})
total_player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items_count = len(purchase_data_df["Item ID"].unique())
avg_price = purchase_data_df["Price"].mean()
purchase_number = len(purchase_data_df["Purchase ID"].unique())
total_price =purchase_data_df["Price"].sum()
total_analysis_df = pd.DataFrame({"Number of Unique Items": [items_count],"Average Price": [avg_price],"Number of Purchases":[purchase_number],"Total Revenue":[total_price]})
total_analysis_df["Average Price"] = total_analysis_df["Average Price"].map("${:.2f}".format)
total_analysis_df["Total Revenue"] = total_analysis_df["Total Revenue"].map("${:.2f}".format)
total_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = purchase_data_df.groupby(["Gender"])
unique_gender= gender_count['SN'].nunique()
gender_count_df = pd.DataFrame((unique_gender),index=['Male','Female','Other / Non-Disclosed'])
percentage_count = pd.DataFrame(round(unique_gender/player_count*100,2))
percentage_count_sign = percentage_count.astype(str) +'%'
percentage_count_df = pd.DataFrame((percentage_count_sign),index=['Male','Female','Other / Non-Disclosed'])
gender_rename = gender_count_df.rename(columns={"SN": "Total Count"})
result = pd.concat([gender_rename,percentage_count_df],axis = 1)
result_rename = result.rename(columns={"SN": "Percentage of Players"})
count_df = pd.DataFrame(result_rename)
count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data_df[["Gender","Purchase ID"]].groupby("Gender").count()
avg_purchase_price =round(purchase_data_df[["Gender","Price"]].groupby("Gender").mean(),2)
total_purchase_price =round(purchase_data_df[["Gender","Price"]].groupby("Gender").sum(),2)
total_purchase_price_rename = total_purchase_price.rename(columns={"Price": "Total Purchase Value"})

result2 = pd.concat([purchase_count,avg_purchase_price,total_purchase_price_rename],axis = 1)
result2_rename = result2.rename(columns={"Purchase ID": "Purchase Count","Price":"Average Purchase Price"})
count2_df = pd.DataFrame(result2_rename)

count_df_gender_count= purchase_data_df[["Gender","SN"]].drop_duplicates(subset = "SN")
gender1_count = count_df_gender_count["Gender"].value_counts(0)
count2_df_gender_count = purchase_data_df.groupby("Gender")
total_purchases= round(count2_df_gender_count["Price"].sum(),2)
count2_df["Avg Total Purchase per Person"]=round(total_purchases/gender1_count,2)

count2_df["Average Purchase Price"] = count2_df["Average Purchase Price"].map("${:.2f}".format)
count2_df["Total Purchase Value"] = count2_df["Total Purchase Value"].map("${:.2f}".format)
count2_df["Avg Total Purchase per Person"] = count2_df["Avg Total Purchase per Person"].map("${:.2f}".format)
count2_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29,34,39,50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True)
age_group = purchase_data_df.groupby("Age Ranges")
age_group = age_group["SN"].nunique()
age_count_df = pd.DataFrame((age_group),index=["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"])
age_count_df_rename =age_count_df.rename(columns={"SN": "Total Count"})
percentage_age_count = pd.DataFrame(round(age_group/player_count*100,2))
percentage_age_sign = percentage_age_count.astype(str) +'%'
result3 = pd.concat([age_count_df_rename ,percentage_age_sign],axis = 1)
result3_rename = result3.rename(columns={"SN": "Percentage of Players"})
count3_df = pd.DataFrame(result3_rename)
count3_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29,34,39,50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True)
age2_group = purchase_data_df.groupby("Age Ranges")["SN"].count()
avg_age_price =round(purchase_data_df[["Age Ranges","Price"]].groupby("Age Ranges").mean(),2)
total_age_price =round(purchase_data_df[["Age Ranges","Price"]].groupby("Age Ranges").sum(),2)
total_age_price_rename = total_age_price.rename(columns={"Price": "Total Purchase Value"})
result4 = pd.concat([age2_group ,avg_age_price,total_age_price_rename],axis = 1)
result4_rename = result4.rename(columns={"SN": "Purchase Count","Price":"Average Purchase Price"})
count4_df = pd.DataFrame(result4_rename)

count3_df_age_count= purchase_data_df[["Age Ranges","SN"]].drop_duplicates(subset = "SN")
age1_count = count3_df_age_count["Age Ranges"].value_counts(0)
count4_df_age_count = purchase_data_df.groupby("Age Ranges")
total_age_purchases= round(count4_df_age_count["Price"].sum(),2)
count4_df["Avg Total Purchase per Person"]=round(total_age_purchases/age1_count,2)

count4_df["Average Purchase Price"] = count4_df["Average Purchase Price"].map("${:.2f}".format)
count4_df["Total Purchase Value"] = count4_df["Total Purchase Value"].map("${:.2f}".format)
count4_df["Avg Total Purchase per Person"] = count4_df["Avg Total Purchase per Person"].map("${:.2f}".format)
count4_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = purchase_data_df[["Item ID", "Item Name","Price"]].groupby(["Item ID","Item Name"])
item_count = item_df["Price"].count().rename("Purchase Count")
avg_item_price = round( item_df["Price"].mean(),2)
total_item_price = round(item_df["Price"].sum(),2)
most_popular_items_df = pd.DataFrame({"Purchase Count":item_count,"Item Price":avg_item_price,"Total Purchase Value":total_item_price})
most_popular_items_df = most_popular_items_df.sort_values("Purchase Count", ascending=False).head()
most_popular_items_df["Item Price"] = most_popular_items_df["Item Price"].map("${:.2f}".format)
most_popular_items_df["Total Purchase Value"] = most_popular_items_df["Total Purchase Value"].map("${:.2f}".format)
most_popular_items_df
```

## Most Profitable Items
